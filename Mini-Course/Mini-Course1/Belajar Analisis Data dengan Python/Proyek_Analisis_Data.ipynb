## Exploratory Data Analysis (EDA)

### Explore ...

```python
correlation_matrix = day_df.corr()

# Visualisasi heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')
plt.show()
```

```python
numerical_columns = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

# Visualisasi distribusi menggunakan histogram dan KDE
plt.figure(figsize=(12, 8))

for column in numerical_columns:
    sns.histplot(day_df[column], kde=True, label=column)

plt.title('Distribusi Kolom Numerik')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='season', y='cnt', data=day_df)
plt.title('Hubungan antara Season dan Count')
plt.xlabel('Season (1:springer, 2:summer, 3:fall, 4:winter)')
plt.ylabel('Count')
plt.show()

# Visualisasi box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=day_df)
plt.title('Hubungan antara Season dan Count')
plt.xlabel('Season (1:spring, 2:summer, 3:fall, 4:winter)')
plt.ylabel('Count')
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
daily_rentals_trend = day_df.groupby('dteday')['cnt'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_rentals_trend.index, y=daily_rentals_trend.values)
plt.title('Tren Harian Penyewaan Sepeda')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.show()

# Tren musiman
seasonal_rentals = hour_df.groupby(['yr', 'season'])['cnt'].mean().unstack()
seasonal_rentals.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Tren Musiman Penyewaan Sepeda (2011-2012)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Penyewaan Rata-rata')
plt.xticks([0, 1], ['2011', '2012'], rotation=0)
plt.legend(title='Musim', labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
```

### Pertanyaan 2:

```python
weather_conditions = ['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow']
plt.figure(figsize=(10, 6))
sns.boxplot(x='weathersit', y='cnt', data=hour_df, palette='coolwarm')
plt.title('Pengaruh Kondisi Cuaca terhadap Jumlah Penyewaan Sepeda')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Penyewaan')
plt.xticks([0, 1, 2, 3], weather_conditions)
plt.show()
```

## Conclusion

Conclution pertanyaan 1
- Tren Harian penyewaan sepeda mengalami peningkatan pada tahun 2012
- Rata-rata penyewaan sepeda tertinggi pada musim gugur di tahun 2011 dan 2012

Conclution pertanyaan 2
- Kondisi cuaca "Clear" memiliki median jumlah penyewaan yang lebih tinggi dibandingkan dengan kondisi cuaca lainnya.
- Kondisi cuaca "Mist" juga memiliki median yang tinggi, tetapi dengan variabilitas yang lebih besar.
- Kondisi cuaca yang cerah ("Clear") dan kondisi cuaca berkabut ("Mist") cenderung memiliki pengaruh positif terhadap jumlah penyewaan sepeda.
- Kondisi cuaca dengan hujan/salju ringan memiliki pengaruh negatif yang lebih nyata, dan kondisi cuaca ekstrem dengan hujan/salju berat memiliki dampak yang paling signifikan dalam mengurangi jumlah penyewaan.
